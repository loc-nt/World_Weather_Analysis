```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part two into a DataFrame.
travel_df = pd.read_csv("WeatherPy_vacation.csv")
travel_df.head()
```

```python
travel_df
```

```python
# Extract 4 cities:
city_A = travel_df.loc[travel_df.City == "San Ramon"].reset_index()
city_B = travel_df.loc[travel_df.City == "Laguna"].reset_index()
city_C = travel_df.loc[travel_df.City == "Jackson"].reset_index()
city_D = travel_df.loc[travel_df.City == "Portland"].reset_index()
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Get location
locA = (city_A.loc[0,'Lat'], city_A.loc[0,'Lng'])
locB = (city_B.loc[0,'Lat'], city_B.loc[0,'Lng'])
locC = (city_C.loc[0,'Lat'], city_C.loc[0,'Lng'])
locD = (city_D.loc[0,'Lat'], city_D.loc[0,'Lng'])
```

```python
# gmap travel:
fig = gmaps.figure()
drive = gmaps.directions_layer(locA, locD, waypoints=[locB, locC], travel_mode='DRIVING')
fig.add_layer(drive)
fig
```

```python
# Combine all 4 cities into 1 df:
combined_df = city_A.append(city_B).append(city_C).append(city_D)
combined_df
```

```python
# Create pop-up markers:
# HTML code:
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in combined_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = combined_df[["Lat", "Lng"]]
fig = gmaps.figure(center=(40.0, -120.0), zoom_level=4.5)

# add pop-up marker:
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig

```
